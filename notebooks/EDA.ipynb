# Exploratory Data Analysis (EDA) for UrbanMart Retail Optimization Analysis Project

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Superstore_Cleaned.csv')
df.info()
```

## 1. Top 10 Best-Selling Products

```python
top_products=df.groupby('Product_Name')['Sales'].sum().sort_values(ascending=False).head(10)
print("Top 10 Best-Selling Products:\n",top_products)
```

```python
#Plot Top 10 Best-Selling Products
top_products.plot(kind='bar',title='Top 10 Best-Selling Products',figsize=(10,5),color='green')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```

## 2. Least Profitable Sub-Categories

```python
loss_makers =df.groupby('Sub_Category')['Profit'].sum().sort_values(ascending=True).head(10)
print("\nLeast Profitable Sub-Categories:\n",loss_makers)
```

```python
#Plot 10 Least Profitable Sub-Categories
loss_makers.plot(kind='bar',title='10 Least Profitable Sub-Categories',figsize=(10,5),color='darkred')
plt.ylabel('Total Profit')
plt.tight_layout()
plt.show()
```

## 3.Sales & Profit by Customer Segment

```python
segment_performance=df.groupby('Segment')[['Sales','Profit']].sum()
print("\nSegment Performance:\n",segment_performance)
```

```python
#Plot
segment_performance.plot(kind='bar',title='Sales & Profit by Customer Segment',figsize=(8,5))
plt.tight_layout()
plt.show()
```

## 4. Regional Sales Performance

```python
region_sales=df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
print("\nSales by Region:\n",region_sales)
```

```python
region_sales.plot(kind='bar',title='Sales by Region',figsize=(8,5),color='midnightblue')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```

## 5. Discounts vs. Profit Analysis

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df,x='Discount',y='Profit')
plt.title('Discount vs. Profit')
plt.tight_layout()
plt.show()
```
